# Mobile App Statistics (Apple iOS app store)

The ever-changing mobile landscape is a challenging space to navigate. . The percentage of mobile over desktop is only increasing. Android holds about 53.2% of the smartphone market, while iOS is 43%. To get more people to download your app, you need to make sure they can easily find your app. Mobile app analytics is a great way to understand the existing strategy to drive growth and retention of future user.

With million of apps around nowadays, the following data set has become very key to getting top trending apps in iOS app store. This data set contains more than 7000 Apple iOS mobile application details. 

#### Data description:

   * "id" : App ID

   * "track_name": App Name

   * "size_bytes": Size (in Bytes)

   * "currency": Currency Type

   * "price": Price amount

   * "ratingcounttot": User Rating counts (for all version)

   * "ratingcountver": User Rating counts (for current version)

   * "user_rating" : Average User Rating value (for all version)

   * "userratingver": Average User Rating value (for current version)

   * "ver" : Latest version code

   * "cont_rating": Content Rating

   * "prime_genre": Primary Genre

   * "sup_devices.num": Number of supporting devices

   * "ipadSc_urls.num": Number of screenshots showed for display

   * "lang.num": Number of supported languages

   * "vpp_lic": Vpp Device Based Licensing Enabled

## Create a new column 'payment_type' and categorize all the applications into 'free' or 'paid'

```python
def is_free(price):
    """
    check if apps are paid or free
    ----
    Params:
    -------
    price: int or float
    
    Returns:
    --------
    free if price is 0 else paid
    """
    assert type(price) in [int, float], "You need to enter a float or an integer"
    assert price >= 0, "Enter a non negative number"
    
    return 'free' if price == 0 else 'paid'
```

```python
is_free(0)
```

```python
is_free(0.05)
```

```python
apple['paid'] = apple['price'].map(is_free)
```

```python
apple.head()
```

```python
sns.countplot(apple['paid']) ## maximum apps are free
```

```python
def price_cat(price):
    if price==0:
        return 'free'
    elif price<4:
        return 'economical'
    elif price<10:
        return 'moderate'
    elif price<50:
        return 'expensive'
    else:
        return 'premium'
```

```python
apple['price_category']=apple['price'].map(price_cat)
```

```python
# Side by side boxplot for price distribution based on category which are paid
apple[apple['price']>0].boxplot(column='price', by='price_category', grid=False)
```

### Create a new column "size_mb" converting bytes to megabytes. For the applications below 50 dollars, does the price of the app increases with the increase in its size?

```python
apple['size_mb'] = apple['size_bytes'].map(lambda x: round(x/(1024**2), 2))
```

## Can we create a recommendation column based on "user_rating" and "rating_count_tot"?

```python
def shortlist(user_rating,rating_count):
           
    return apple[(apple['user_rating']>user_rating) & (apple['rating_count_tot']>rating_count)]['track_name'].values
```

```python
myapps = shortlist(4.2,2100)
```

```python
## price vs user rating.
price_rating = apple.groupby(['user_rating','paid']).size().unstack()
ax= price_rating.plot(kind='bar', stacked=True, figsize=(15,10))
plt.xlabel('user_rating')
plt.ylabel('paid')
plt.xticks(rotation=45)
plt.show()
#Their is no such relation between price and user_rating. Most of the free apps has higher user _rating of 4.5.
```

```python
price_gen =apple.groupby(['prime_genre','paid']).size().unstack()
ax= price_gen.plot(kind='bar',stacked= True, figsize=(15,10))
plt.xlabel('prime_genre')
plt.ylabel('paid')
plt.xticks(rotation=45)
plt.show()
```
